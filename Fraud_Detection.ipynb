```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

Data Collection and processing

```python
#loading the dataset to a pandaas dataframe
credit_card_data=pd.read_csv('/content/creditcard.csv')
```

```python
# Print the first five rows of the DataFrame
credit_card_data.head()
```

```python

# Print the last five rows of the dataframe
credit_card_data.tail()
```

```python
#dataset info
credit_card_data.info()
```

```python
#checking the no. of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transaction and froudlent transaction
credit_card_data['Class'].value_counts()
```

 this dataset is highly imbalanced
 0-->normal transaction
 1-->fraudlent transection

```python
# seperating the data for analysis
legit = credit_card_data[credit_card_data.Class==0]
fraud= credit_card_data[credit_card_data.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
print(legit.Amount.describe())
print(fraud.Amount.describe())
```

```python
#compare the values for both transaction
credit_card_data.groupby('Class').mean()
```

 undersampling

```python
# build a sample data set containing similar distribution of normal transaction and fraudlent transaction
# no. of fraudlent transaction->471
legit_sample = legit.sample(n=471)#randomly select data from legit dataset
```

```python
#concatenating two data frame
# axis=0 means concat rowwisw
# axis=1 means concat columnwise
new_dataset = pd.concat([legit_sample,fraud],axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# splitting the data into features and targets
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
print(X)
print(Y)
```

```python
# split the data into training data and testing data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
# here 80% of the data goes to training and 20% of the data goes for testing
print(X.shape,X_train.shape,X_test.shape)
```

model training(logistic regression model-->for binary classification)

```python
model=LogisticRegression()
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
#training the logistic regression model with training data
model.fit(X_train_scaled, Y_train)
```

evaluation of the model based on the accuracy score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train_scaled)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print('Accuracy on Training DaTA:', training_data_accuracy)
```

```python
#accuracy on testing data
model.fit(X_test_scaled,Y_test)
X_test_prediction = model.predict(X_test_scaled)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print('Accuracy on Testing Data:',test_data_accuracy)
```
